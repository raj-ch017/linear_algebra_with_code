## 1.1 - Vectors : Geometric Representation

A vector is a line, which is determined by a magnitude and a direction. Lines can exist on a coordinate system with any number of dimensions greater than zero. The dimensionality of the vector is the dimensionality of the coordinate system in which it is defined.

It is **important** to know that the definition of a vector does not include its starting or ending locations. That is, the vector `[1 -2]` simply means a line that goes one unit in the positive direction in the first dimension and two units in the negative direction in the second dimension.

For any coordinate system, a given coordinate is an unique point in space. A vector, however, is any line for which the end point (also called the _head_ of the vector) is a certain number of units along each dimension away from the starting point (also known as _tail_ of the vector).

On the other hand, coordinates and vectors are coincident when the vector starts at the origin. A vector with it's tail at the origin is said to be in it's _standard position_.

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
vector1 = np.array([1, -2])

plt.figure(figsize= (8, 8))
plt.plot([0, vector1[0]], [0, vector1[1]])
plt.axis([-3, 3, -3, 3])
plt.plot(0,0, "ro", label = "Origin - Vector Tail")
plt.plot(1, -2, "go", label = "Coordinate - Vector Head")
plt.title("2D Vector in Standard Position")
plt.xlabel("$X_{1} dim$")
plt.ylabel("$X_{2} dim$")
plt.legend();
```

```python
vector2 = np.array([3, 1, 4])

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()

ax.plot3D([0, vector2[0]], [0, vector2[1]], [0, vector2[2]])
ax.set_title('3D Vector in Standard Position')

ax.plot3D(0,0,0,'ro', label = "origin")
ax.plot3D(3,1,4, 'go', label = "head")

ax.set_xlabel("$X_{1} dim$", labelpad=20)
ax.set_ylabel("$X_{2} dim$", labelpad=20)
ax.set_zlabel("$X_{3} dim$", labelpad=10)

plt.legend()
```

## 1.2 - Vectors: Algebraic Representation

Vector is an ordered collection of numbers, known as the components of the vector. The dimensionality of a vector is the number of components present inside it.

* **2 dimensional vector**: `[1 2]`
* **3 dimensional vector**: `[1 3 7]`
* **4 dimensional vector**: `[-3 0 1 8]`

The ordering of the numbers is important. **For example**, the following two vectors are not the same, even though they have the same dimensionality and the same components.

- `[1 3]` and `[3 1]`

Vectors are not limited to collecting numbers. The elements of the vector can be functions as well. Consider the vector:

        `[cos(t) sin(t) t]`

where, t is a vector of time points.

```python
# Setting up figure and grid
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()

# Defining vector and functions
t = np.arange(0, 10 * np.pi, np.pi / 50)
x = np.cos(t)
y = np.sin(t)

# Plotting the figure
ax.plot3D(x, y, t)
ax.set_title('3D Plot of [cos(t) sin(t) t]')

# Set axes label
ax.set_xlabel('$sin(t)$', labelpad=20)
ax.set_ylabel('$cos(t)$', labelpad=20)
ax.set_zlabel('t', labelpad=10)

plt.show();
```

### `Zeros Vector`

The vector that contains zeros in all of its elements is called the zeros vector. A vector that contains some zeros but other non-zero elements is not the zeros vector; it's just a regular vector with some zero components.

The zeros vector has some interesting and weird properties. One of these properties is that it doesn't have a direction. In other words, it's direction is undefined. That's because the zeros vector is simply a point at the origin of a graph. Without any magnitude, it doesn't make sense to ask which direction it points in.

Create a 2D vector **v**, 10 scalars that are drawn at random from a normal (Gaussian) distribution, and plot all 10 scalar-vector multiplications on top of each other.

```python
random_scalars = [np.random.normal() for x in range(10)]

v = np.array([1, 2])
```

```python
plt.figure(figsize=(8, 8))
num = 1

for scalar in random_scalars:

    new_vector = scalar * v 
    plt.plot([0,new_vector[0]], [0, new_vector[1]], label = "vector" + str(num))
    num += 1

plt.grid('on')
plt.axis([-5, 5, -5, 5])
plt.legend();
```
